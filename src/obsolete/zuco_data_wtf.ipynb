```python
def load_and_inspect_feature_file(filepath):
    """
    Load and inspect the contents of a feature file
    """
    print(f"File exists: {os.path.exists(filepath)}")
    print(f"File size: {os.path.getsize(filepath) / 1024:.2f} KB")
    
    # Load with pickle support
    data = np.load(filepath, allow_pickle=True)
    
    # Basic information
    print(f"\nData type: {type(data)}")
    print(f"Data shape: {data.shape if hasattr(data, 'shape') else 'No shape'}")
    print(f"Data dtype: {data.dtype if hasattr(data, 'dtype') else 'No dtype'}")
    
    # Try to inspect contents based on type
    if isinstance(data, np.ndarray):
        if data.dtype == np.dtype('O'):  # Object array
            print("\nThis is an object array. Let's inspect the first element:")
            first_item = data.item() if data.size == 1 else data[0]
            print(f"First item type: {type(first_item)}")
            
            # If dictionary, show keys
            if isinstance(first_item, dict):
                print(f"Dictionary keys: {list(first_item.keys())}")
                # Show a sample of values
                for key in list(first_item.keys())[:3]:  # First 3 keys
                    value = first_item[key]
                    print(f"  Key '{key}': {type(value)}, ", end="")
                    if hasattr(value, 'shape'):
                        print(f"shape {value.shape}")
                    else:
                        print(f"value {value}")
    
    return data
```

```python
def load_and_organize_eeg_features(filepath):
    """
    Load EEG features and organize them by subject, task, and sentence block
    """
    # Load the data
    data = np.load(filepath, allow_pickle=True).item()
    
    # Create a structured dictionary
    organized_data = {}
    
    for key, value in data.items():
        parts = key.split('_')
        subject_id = parts[0]
        task = parts[1]
        block_id = parts[2]
        index = parts[3]
        
        # Extract EEG features (excluding task label)
        features = np.array(value[:-1])
        
        # Organize by subject  task  block  index
        if subject_id not in organized_data:
            organized_data[subject_id] = {}
        
        if task not in organized_data[subject_id]:
            organized_data[subject_id][task] = {}
        
        if block_id not in organized_data[subject_id][task]:
            organized_data[subject_id][task][block_id] = {}
        
        organized_data[subject_id][task][block_id][index] = features
    
    return organized_data

def extract_word_features(features_dict):
    """Extract word-level EEG features from the dictionary"""
    features = []
    word_metadata = []
    
    for key, value in features_dict.items():
        parts = key.split('_')
        subject_id = parts[0]
        task = parts[1]
        sentence_id = parts[2]
        word_idx = parts[3]
        
        # All elements except the last (task label)
        eeg_features = np.array(value[:-1])
        
        features.append(eeg_features)
        word_metadata.append({
            'key': key,
            'subject': subject_id,
            'task': task,
            'sentence_id': sentence_id,
            'word_idx': word_idx
        })
    
    return np.array(features), word_metadata
```

```python
import numpy as np
import os

def load_word_level_eeg_features(filepath):
    """
    Load word-level EEG features from the .npy file.
    
    Parameters:
    -----------
    filepath : str
        Path to the .npy file containing EEG features
    
    Returns:
    --------
    dict
        Dictionary mapping word IDs to EEG features
    """
    # Check if file exists
    if not os.path.exists(filepath):
        raise FileNotFoundError(f"File not found: {filepath}")
    
    # Load the data with allow_pickle=True
    data = np.load(filepath, allow_pickle=True)
    
    # Extract the dictionary from the object array
    if isinstance(data, np.ndarray) and data.dtype == np.dtype('O'):
        if data.size == 1:
            features_dict = data.item()
            if isinstance(features_dict, dict):
                return features_dict
    
    # If we get here, the file doesn't contain what we expected
    raise ValueError("The file doesn't contain a dictionary in the expected format")

def analyze_word_features(features_dict):
    """
    Analyze the structure of the word-level EEG features.
    
    Parameters:
    -----------
    features_dict : dict
        Dictionary of word-level EEG features
    
    Returns:
    --------
    dict
        Dictionary with analysis results
    """
    # Get some basic stats
    num_words = len(features_dict)
    sample_key = next(iter(features_dict))
    sample_value = features_dict[sample_key]
    
    # Count NR and TSR words
    nr_count = sum(1 for key in features_dict if '_NR_' in key)
    tsr_count = sum(1 for key in features_dict if '_TSR_' in key)
    
    # Extract unique subject IDs
    subject_ids = set(key.split('_')[0] for key in features_dict)
    
    # Extract unique sentence IDs
    sentence_ids = set(f"{key.split('_')[0]}_{key.split('_')[1]}_{key.split('_')[2]}" 
                      for key in features_dict)
    
    # Get feature dimensionality (number of electrodes)
    # Subtract 1 for the task label at the end
    num_electrodes = len(sample_value) - 1
    
    # Return analysis results
    return {
        'num_words': num_words,
        'nr_count': nr_count,
        'tsr_count': tsr_count,
        'subject_ids': list(subject_ids),
        'num_sentences': len(sentence_ids),
        'num_electrodes': num_electrodes,
        'sample_key': sample_key,
        'sample_value_length': len(sample_value),
        'sample_value_slice': sample_value[:5],  # First 5 electrode values
        'task_label': sample_value[-1]  # Last element is the task label
    }

def prepare_word_data_for_mapping(features_dict):
    """
    Prepare the word-level EEG data for mapping with a language model.
    
    Parameters:
    -----------
    features_dict : dict
        Dictionary of word-level EEG features
    
    Returns:
    --------
    tuple
        X: numpy array of shape (n_words, n_electrodes)
        word_info: list of dictionaries with word metadata (subject_id, task, sentence_id, word_position)
    """
    # Initialize lists to store data
    X = []
    word_info = []
    
    for key, value in features_dict.items():
        # Extract EEG features (all but the last element)
        features = np.array(value[:-1])
        
        # Parse key to extract metadata
        parts = key.split('_')
        subject_id = parts[0]
        task = parts[1]
        sentence_id = parts[2]
        word_position = parts[3]
        
        # Append to lists
        X.append(features)
        word_info.append({
            'subject_id': subject_id,
            'task': task,
            'sentence_id': sentence_id,
            'word_position': int(word_position),
            'key': key
        })
    
    # Convert to numpy array
    X = np.array(X)
    
    return X, word_info
```
